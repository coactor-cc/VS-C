```python
import pandas as pd
```

```python
# 读取数据集
file_path = r"..\data\dataset.csv"
data = pd.read_csv(file_path)
```

```python
# 性别字段
gender_values = data['gender'].unique()
print(f"gender: {', '.join(gender_values)}")

# 年龄字段
age_min = data['age'].min()
age_max = data['age'].max()
print(f"age: [{age_min}-{age_max}]")

# 高血压字段
hypertension_values = data['hypertension'].unique()
print(f"hypertension: {', '.join(map(str, hypertension_values))}")

# 心脏病字段
heart_disease_values = data['heart_disease'].unique()
print(f"heart_disease: {', '.join(map(str, heart_disease_values))}")

# 婚姻状态字段
ever_married_values = data['ever_married'].unique()
print(f"ever_married: {', '.join(ever_married_values)}")

# 工作类型字段
work_type_values = data['work_type'].unique()
print(f"work_type: {', '.join(work_type_values)}")

# 居住类型字段
residence_type_values = data['Residence_type'].unique()
print(f"Residence_type: {', '.join(residence_type_values)}")

# 平均血糖水平字段
avg_glucose_min = data['avg_glucose_level'].min()
avg_glucose_max = data['avg_glucose_level'].max()
print(f"avg_glucose_level: [{avg_glucose_min}-{avg_glucose_max}]")

# 体重指数字段
bmi_min = data['bmi'].min()
bmi_max = data['bmi'].max()
print(f"bmi: [{bmi_min}-{bmi_max}]")

# 吸烟状态字段
# smoking_status_values = data['smoking_status'].unique()
# print(f"smoking_status: {', '.join(smoking_status_values)}")

# 中风字段
stroke_values = data['stroke'].unique()
print(f"stroke: {', '.join(map(str, stroke_values))}")
```
